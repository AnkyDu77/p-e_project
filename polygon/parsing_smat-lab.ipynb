```python
import requests
from bs4 import BeautifulSoup as bs
import re

url = 'https://smart-lab.ru/q/shares/'
req = requests.get(url)
soup = bs(req.text)
soup_str = str(soup)

with open('text_sl.txt', 'w') as file:
    file.write(soup_str)
```

```python
res_link_lst = []
with open('text_sl.txt', 'r') as file_prs:
    for line in file_prs:
        item = re.findall('/forum.+?\s', line)
        res_link_lst.append(item)
```

```python
clean_lst = [value for value in res_link_lst if value != []]
for i in clean_lst:
    print(i)
len(clean_lst)
```

```python
name_lst = []
for item in clean_lst:
    name = re.findall('>[А-Яа-яa-zA-Z0-9-.+]+.<?', item[0])
    name_lst.append(name)
clean_name_lst0 = [value for value in name_lst if value != []]
clean_name_lst1 = []
for item in clean_name_lst0:
    name = item[0]
    clean_name = name[1:-1]
    clean_name_lst1.append(clean_name)
del clean_name_lst1[:2]
del clean_name_lst1[-3:]
for i in clean_name_lst1:
    print(i)
len(clean_name_lst1)
```

```python
link_lst = []
for item in clean_lst:
    link = re.findall('/forum/[A-Za-z0-9%-]*.>', item[0])
    link_lst.append(link)
clean_link_lst0 = [value for value in link_lst if value != []]
clean_link_lst1 = [] 
for i in clean_link_lst0:
    link = i[0]
    clean_link_lst1.append(link)
clean_link_lst1 = clean_link_lst1[1:-1]
clean_link_lst2 = []
for item in clean_link_lst1:
    clean_link = item[:-2]
    clean_link_lst2.append(clean_link)

clean_link_lst3 = []
for i in clean_link_lst2:
    link = re.findall('%[A-Z0-9%]*.', i) #!!!
    clean_link_lst3.append(link)
clean_link_lst4 = [value for value in clean_link_lst3 if value != []]
clean_link_lst4
```
